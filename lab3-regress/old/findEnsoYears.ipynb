```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#NOAA ERSSTv5 data: https://www.esrl.noaa.gov/psd/data/gridded/data.noaa.ersst.v5.html
#MON/MMM    : month for Nino34 values, 3-month season centered on that month for ONI
#NINO34_MEAN: monthly average SST (°C) in Nino3.4 region (5S-5N,190E-240E)
#NINO34_CLIM: climatological 30-year monthly average SST, updated every 5 years
#NINO34_ANOM: difference between the monthly mean and the climatological value
#ONI        : 3-month average anomaly (°C) centered on given month (01=DJF,08=JAS,etc)
#PHASE      : flags ONI value as Missing (M), El Nino (E), Neutral (N), La Nina (L)
#Data table generated by Brian McNoldy, University of Miami (bmcnoldy@rsmas.miami.edu)
# obtained https://bmcnoldy.rsmas.miami.edu/tropics/oni/ONI_NINO34_1854-2021.txt
```

```python
year,mon,nino34_mean,nino34_clim,nino34_anom,oni,phase = np.genfromtxt("monthlyONI.dat",dtype='unicode').T
```

```python
data = pd.DataFrame({'YEAR':year,
              'MONTH':mon,
              'NINO34_MEAN':nino34_mean,
              'NINO34_CLIM':nino34_clim,
              'NINO34_ANOM':nino34_anom,
              'ONI':oni,
              'PHASE':phase})
data.to_csv("nino.csv",index=False) # save 
```
