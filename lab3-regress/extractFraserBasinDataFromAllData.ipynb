```python
import numpy as np
import pandas as pd
```

```python
snow = pd.read_csv("snowWaterEquivalentBC_Apr1.csv")
fraserStations = pd.read_csv("../snow/fraserStations.dat",skiprows=3).stations.values

shared = []
for f in fraserStations:
    if f in snow.Station.values.astype('str'):
        shared.append(f)

mask = []
for s in snow.Station.values.astype('str'):
    if s in shared:
        mask.append(True)
    else:
        mask.append(False)
mask = np.array(mask)


fraserData = snow[mask]

fraserData.Date=fraserData.Date.str.slice(start=0,stop=4)

fraserData=fraserData.rename(columns={'Station':'station','Date':'year','SnowWaterEqui(mm)':'swe', 
                                     'Latitude':'lat','Longitude':'lon','Elevation(m)':'ele'})

fraserData.to_csv('../snow/fraserBasinSnowData.csv',index=False)
```
